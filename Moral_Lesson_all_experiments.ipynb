```python
import os
import csv
import math
import string
from collections import Counter
import re
import warnings
warnings.filterwarnings("ignore")

# datascience libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

# sklearn library
from sklearn.datasets import load_files
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

# algorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
```

```python
para_level_Dataset_features=pd.read_csv('/content/drive/MyDrive/Research/Final Experimentation Datasets/Topic wise advanced/Moral Lesson/para_stylometry_features.csv')
```

```python
para_level_Dataset_features
```

```python
# drop column of file names
para_level_Dataset_features = para_level_Dataset_features.drop(['File Names'], axis=1)
```

```python
# convert Para-Level Label column values into digits
label_encoder = LabelEncoder()
para_level_Dataset_features['Label'] = label_encoder.fit_transform(para_level_Dataset_features['Label'])
```

```python
#all features
X_all=para_level_Dataset_features.drop(['Label'], axis=1)
y=para_level_Dataset_features['Label']
X_all_train, X_all_test, y_train, y_test = train_test_split(X_all, y, test_size=0.2, random_state=42)
# Convert X_train and X_test to NumPy arrays
X_all_train_np = np.array(X_all_train)
X_all_test_np = np.array(X_all_test)

# Ensure the arrays are contiguous
X_all_train_np = np.ascontiguousarray(X_all_train_np)
X_all_test_np = np.ascontiguousarray(X_all_test_np)
```

```python
#word level
X_word=para_level_Dataset_features.iloc[:, 0:10].values
X_word_train, X_word_test, y_train, y_test = train_test_split(X_word, y, test_size=0.2, random_state=42)
# Convert X_train and X_test to NumPy arrays
X_word_train_np = np.array(X_word_train)
X_word_test_np = np.array(X_word_test)

# Ensure the arrays are contiguous
X_word_train_np = np.ascontiguousarray(X_word_train_np)
X_word_test_np = np.ascontiguousarray(X_word_test_np)
```

```python
#Character level
X_char=para_level_Dataset_features.iloc[:, 10:26].values
X_char_train, X_char_test, y_train, y_test = train_test_split(X_char, y, test_size=0.2, random_state=42)
# Convert X_train and X_test to NumPy arrays
X_char_train_np = np.array(X_char_train)
X_char_test_np = np.array(X_char_test)

# Ensure the arrays are contiguous
X_char_train_np = np.ascontiguousarray(X_char_train_np)
X_char_test_np = np.ascontiguousarray(X_char_test_np)
```

```python
#Sen level
X_sen=para_level_Dataset_features.iloc[:, 26:33].values
X_sen_train, X_sen_test, y_train, y_test = train_test_split(X_sen, y, test_size=0.2, random_state=42)
# Convert X_train and X_test to NumPy arrays
X_sen_train_np = np.array(X_sen_train)
X_sen_test_np = np.array(X_sen_test)

# Ensure the arrays are contiguous
X_sen_train_np = np.ascontiguousarray(X_sen_train_np)
X_sen_test_np = np.ascontiguousarray(X_sen_test_np)
```

```python
#para+doc+tex
X_oth=para_level_Dataset_features.iloc[:, 33:].values
X_oth_train, X_oth_test, y_train, y_test = train_test_split(X_oth, y, test_size=0.2, random_state=42)
# Convert X_train and X_test to NumPy arrays
X_oth_train_np = np.array(X_oth_train)
X_oth_test_np = np.array(X_oth_test)

# Ensure the arrays are contiguous
X_oth_train_np = np.ascontiguousarray(X_oth_train_np)
X_oth_test_np = np.ascontiguousarray(X_oth_test_np)
```

##### 1 Apply Machine Learning Algorithms on  Para-Level Dataset

```python
# Initialize classifiers
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
rf = RandomForestClassifier()
svm = SVC()
knn = KNeighborsClassifier()
nb = GaussianNB()
lda = LinearDiscriminantAnalysis()
mlp = MLPClassifier(max_iter=1000)  # Adjust max_iter based on convergence
xgb = GradientBoostingClassifier()
```

```python
# Create a voting classifier
voting_clf = VotingClassifier(estimators=[
    ('logreg', logreg),
    ('dtree', dtree),
    ('rf', rf),
    ('svm', svm),
    ('knn', knn),
    ('nb', nb),
    ('lda', lda),
    ('mlp', mlp),
    ('xgb', xgb)
], voting='hard')
```

##### 1.1 word level

```python
#training
classifiers = [logreg, dtree, rf, svm, knn, nb, lda, mlp, xgb, voting_clf]

for clf in classifiers:
    clf.fit(X_word_train_np, y_train)
```

```python
#testing
classifiers = [logreg, dtree, rf, svm, knn, nb, lda, mlp, voting_clf]
y_pred_all = []

for clf in classifiers:
    y_pred = clf.predict(X_word_test_np)
    y_pred_all.append(y_pred)
```

```python
from sklearn.metrics import confusion_matrix

classifiers = [logreg, dtree, rf, svm, knn, nb, lda, mlp, xgb, voting_clf]

for clf in classifiers:
    # Fit the classifier to the training data
    clf.fit(X_word_train_np, y_train)

    # Make predictions on the test data
    y_pred = clf.predict(X_word_test_np)

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Print the confusion matrix for each classifier
    print(f"Confusion matrix for {clf.__class__.__name__}:\n{cm}\n")
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

classifiers = [logreg, dtree, rf, svm, knn, nb, lda, mlp, xgb, voting_clf]

for clf in classifiers:
    # Fit the classifier to the training data
    clf.fit(X_word_train_np, y_train)

    # Make predictions on the test data
    y_pred = clf.predict(X_word_test_np)

    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Print metrics with bold headings
    print(f"\nMetrics for {clf.__class__.__name__}:")
    print(f"**Accuracy:** {accuracy:.4f}")
    print(f"**Recall:** {recall:.4f}")
    print(f"**Precision:** {precision:.4f}")
    print(f"**F1 Score:** {f1:.4f}")
```
