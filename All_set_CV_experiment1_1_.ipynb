```python
import os
import csv
import math
import string
from collections import Counter
import re
import warnings
warnings.filterwarnings("ignore")

# datascience libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

# sklearn library
from sklearn.datasets import load_files
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_score, KFold
# algorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
```

```python
para_level_Dataset_features=pd.read_csv('/content/drive/MyDrive/Research/Final codes/all_stylometry_features (2).csv')
```

```python
# drop column of file names
para_level_Dataset_features = para_level_Dataset_features.drop(['File Names'], axis=1)
```

```python
# convert Para-Level Label column values into digits
label_encoder = LabelEncoder()
para_level_Dataset_features['Label'] = label_encoder.fit_transform(para_level_Dataset_features['Label'])
```

```python
#all features
X_all=para_level_Dataset_features.drop(['Label'], axis=1)
y=para_level_Dataset_features['Label']
```

```python
#word level
X_word=para_level_Dataset_features.iloc[:, 0:10].values
```

```python
#Character level
X_char=para_level_Dataset_features.iloc[:, 10:26].values
```

```python
#Sen level
X_sen=para_level_Dataset_features.iloc[:, 26:33].values
```

```python
#para+doc+tex
X_oth=para_level_Dataset_features.iloc[:, 33:].values
```

```python
# Initialize classifiers
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
rf = RandomForestClassifier()
svm = SVC()
knn = KNeighborsClassifier()
nb = GaussianNB()
lda = LinearDiscriminantAnalysis()
mlp = MLPClassifier(max_iter=1000)  # Adjust max_iter based on convergence
xgb = GradientBoostingClassifier()
```

```python
# Create a voting classifier
voting_clf = VotingClassifier(estimators=[
    ('logreg', logreg),
    ('dtree', dtree),
    ('rf', rf),
    ('svm', svm),
    ('knn', knn),
    ('nb', nb),
    ('lda', lda),
    ('mlp', mlp),
    ('xgb', xgb)
], voting='hard')
```

```python
# Initialize classifiers list
classifiers = [logreg, dtree, rf, svm, knn, nb, lda, mlp, xgb, voting_clf]
```

```python
# Define number of folds
n_folds = 5

# Initialize KFold object
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
```

**All features**

```python
# Iterate over classifiers
for clf in classifiers:
    # Perform k-fold cross-validation
    scores_accuracy = cross_val_score(clf, X_all, y, cv=kf, scoring='accuracy')
    scores_precision = cross_val_score(clf, X_all, y, cv=kf, scoring='precision')
    scores_recall = cross_val_score(clf, X_all, y, cv=kf, scoring='recall')
    scores_f1 = cross_val_score(clf, X_all, y, cv=kf, scoring='f1')

    # Calculate mean scores
    mean_accuracy = np.mean(scores_accuracy)
    mean_precision = np.mean(scores_precision)
    mean_recall = np.mean(scores_recall)
    mean_f1 = np.mean(scores_f1)

    # Print mean scores
    print(f"Classifier: {clf.__class__.__name__}")
    print(f"Mean Accuracy: {mean_accuracy:.4f}")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1 Score: {mean_f1:.4f}")
    print("------------------------------")
```
